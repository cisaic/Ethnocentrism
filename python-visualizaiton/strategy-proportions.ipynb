```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

# Strategy proportions by tag for 10x10 and 50x50

```python
df = pd.read_csv("strategy-proportions.csv", sep=",", header=None)
df.columns = df.iloc[0]
df = df.reindex(df.index.drop(0))
#df.size = pd.to_numeric(df.size, errors='coerce')
df.tag = pd.to_numeric(df.tag, errors='coerce')
df.proportion = pd.to_numeric(df.proportion, errors='coerce')
df.head()
```

```python
for size, df2 in df.groupby('size'):
    plt.figure(figsize=(7,7))
    sns.boxplot(x='tag', y='proportion', hue='strategy', data=df2)
    plt.title("strategy proportions by tag for pop length = " + str(size))
    plt.ylim(top=1.0)
```
