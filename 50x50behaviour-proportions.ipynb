```python
# 50x50 2 tags

n50_tags2_bp = pd.read_csv(file50_2_run[0], sep="\t", header=None)
n50_tags2_bp = n50_tags2_bp.drop(2, axis=1)
n50_tags2_bp.columns = ["cooperate", "defect"]

#worlds = []
#i = 1
#for row in n10_tags2_bp['cooperate']:
#    worlds.append(i)
#    i = i + 1
#n10_tags2_bp.insert(loc=0, column='world', value=worlds)   

n50_tags2_bp["size"] = 50
n50_tags2_bp["tag"] = 2
n50_tags2_bp
```

```python
# 50x50 4 tags
n50_tags4_bp = pd.read_csv(file50_4_run[0], sep="\t", header=None)
n50_tags4_bp = n50_tags4_bp.drop(2, axis=1)
n50_tags4_bp.columns = ["cooperate", "defect"]
n50_tags4_bp["size"] = 50
n50_tags4_bp["tag"] = 4
```

```python
# 50x50 8 tags
n50_tags8_bp = pd.read_csv(file50_8_run[0], sep="\t", header=None)
n50_tags8_bp = n50_tags8_bp.drop(2, axis=1)
n50_tags8_bp.columns = ["cooperate", "defect"]
n50_tags8_bp["size"] = 50
n50_tags8_bp["tag"] = 8
```

```python
# 50x50 1k tags
n50_tags10k_bp = pd.read_csv(file50_10k_run[0], sep="\t", header=None)
n50_tags10k_bp = n50_tags10k_bp.drop(2, axis=1)
n50_tags10k_bp.columns = ["cooperate", "defect"]
n50_tags10k_bp["size"] = 50
n50_tags10k_bp["tag"] = 1000
```

```python
n50_bp = n50_tags1_bp
n50_bp = n50_bp.append(n50_tags2_bp)
n50_bp = n50_bp.append(n50_tags4_bp)
n50_bp = n50_bp.append(n50_tags8_bp)
n50_bp = n50_bp.append(n50_tags10k_bp)
```

```python
n50_bp_m = pd.melt(n50_bp, id_vars=['tag'], value_vars=['cooperate', 'defect'], var_name='behaviour', value_name='proportion')
n50_bp_m
```

```python
# 50x50 box plot behaviour proportions by tag
plt.figure(figsize=(7,7))
sns.boxplot(x='tag', y='proportion', hue='behaviour', data=n50_bp_m)
plt.title("50x50 behaviour proportions by tag")
plt.ylim(top=1.0)
```
