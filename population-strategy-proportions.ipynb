```python
# 10x10 2 tags

n10_tags2_sp = pd.read_csv(file10_2_run[3], sep="\t", header=None)
n10_tags2_sp = n10_tags2_sp.drop(4, axis=1)
n10_tags2_sp.columns = ["selfish", "traitor","ethnocentric","humanitarian"]
n10_tags2_sp["size"] = 10
n10_tags2_sp["tag"] = 2
n10_tags2_sp
```

```python
# 10x10 4 tags

n10_tags4_sp = pd.read_csv(file10_4_run[3], sep="\t", header=None)
n10_tags4_sp = n10_tags4_sp.drop(4, axis=1)
n10_tags4_sp.columns = ["selfish", "traitor","ethnocentric","humanitarian"]
n10_tags4_sp["size"] = 10
n10_tags4_sp["tag"] = 4
n10_tags4_sp
```

```python
# 10x10 8 tags

n10_tags8_sp = pd.read_csv(file10_8_run[3], sep="\t", header=None)
n10_tags8_sp = n10_tags8_sp.drop(4, axis=1)
n10_tags8_sp.columns = ["selfish", "traitor","ethnocentric","humanitarian"]
n10_tags8_sp["size"] = 10
n10_tags8_sp["tag"] = 8
n10_tags8_sp
```

```python
# 10x10 1k tags

n10_tags1k_sp = pd.read_csv(file10_1k_run[3], sep="\t", header=None)
n10_tags1k_sp = n10_tags1k_sp.drop(4, axis=1)
n10_tags1k_sp.columns = ["selfish", "traitor","ethnocentric","humanitarian"]
n10_tags1k_sp["size"] = 10
n10_tags1k_sp["tag"] = 10000
n10_tags1k_sp
```

```python
n10_sp = n10_tags1_sp
n10_sp = n10_sp.append(n10_tags2_sp)
n10_sp = n10_sp.append(n10_tags4_sp)
n10_sp = n10_sp.append(n10_tags8_sp)
n10_sp = n10_sp.append(n10_tags1k_sp)
```

```python
n10_sp_m = pd.melt(n10_sp, id_vars=['size','tag'], value_vars=['selfish', 'traitor','ethnocentric','humanitarian'], var_name='strategy', value_name='proportion')
n10_sp_m
```

```python
# 10x10 box plot strategy proportions by tag
plt.figure(figsize=(7,7))
sns.boxplot(x='tag', y='proportion', hue='strategy', data=n10_sp_m)
plt.title("10x10 strategy proportions by tag")
plt.ylim(top=1.0)
```
