```python
# 50x50 tags 1 
n50_tags1 = pd.read_csv(file50_1_runall[0], sep="\t", header=None)
n50_tags1 = n50_tags1.drop(5, axis=1)
n50_tags1 = n50_tags1.drop(4, axis=1)
n50_tags1.columns = ["selfish", "traitor","ethnocentric","humanitarian"]
n50_tags1["size"] = 50
n50_tags1["tag"] = 1
n50_tags1["world"] = 1

# Create array to index number of cycles
cycles = []
i = 1
for row in n50_tags1['tag']:
    cycles.append(i)
    i = i + 1
    
n50_tags1.insert(loc=0, column='cycle', value=cycles)

n50_tags1.head()
```

```python
# all worlds for 10x10 population, tags = 1
world = 2
for file in file50_1_runall[1:]:
    
    temp = pd.read_csv(file, sep="\t", header=None)
    temp = temp.drop(5, axis=1)
    temp = temp.drop(4, axis=1)
    temp.columns = ["selfish", "traitor","ethnocentric","humanitarian"]
    temp["size"] = 50
    temp["tag"] = 1
    temp["world"] = world
    temp.insert(loc=0, column='cycle', value=cycles)
    
    world += 1
    n50_tags1 = n50_tags1.append(temp)
```
